````python
from sklearn.model_selection import StratifiedShuffleSplit
import pandas as pd

# Example dataset
data = {
    "Feature1": [1, 2, 3, 4, 5],
    "Feature2": [5, 4, 3, 2, 1],
    "Feature3": [15, 14, 13, 12, 11],
    "Feature4": [11, 12, 13, 14, 15],
    "Feature5": [10, 9, 8, 7, 6],
    "Feature6": [6, 7, 8, 9, 10],
    "stratum_column": [0, 0, 1, 1, 1]  # Imbalanced classes
}
df = pd.DataFrame(data)

split = StratifiedShuffleSplit(n_splits=2, test_size=0.5, random_state=42)
for i,(train_index, test_index) in enumerate(split.split(df, df["stratum_column"])):
        print(f"Split {i + 1}:")
        print(train_index)
        print(test_index)
        strat_train_set = df.loc[train_index]
        strat_test_set = df.loc[test_index]
        print("\nTraining Set:")
        print(strat_train_set)
        print("\nTest Set:")
        print(strat_test_set)
        print("-" * 40)




````

# Rolling Cross-Validation
n_splits = 5
window_size = 2  # Size of the test set
for i in range(n_splits):
    train_end = len(X) - (n_splits - i) * window_size
    train_indices = np.arange(0, train_end)
    test_indices = np.arange(train_end, train_end + window_size)

    X_train, X_test = X[train_indices], X[test_indices]
