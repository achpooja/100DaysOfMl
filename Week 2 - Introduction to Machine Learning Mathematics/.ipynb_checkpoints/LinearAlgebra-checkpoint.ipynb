Why Vector Addition and Subtraction matter?

 - Feature Engineering

Machine Learning models often benefit from well-designes features. Vectors addition and subtraction are cruicial for feature enginnering. For example, in image processing, subtracting background color vectors from images can enhance object recognition, making it easier to identify handwritten digits in optical character recognition (OCR).

For any vectors, whether it is x, y, z(of the same size/shape)

-  The following properties are true:
 -  Vector addition are commutative: x+y=y+x
 - Vector addition are associative:(x+y)+z = x+(y+z). Therefore x+y+z is true.
 - Adding a Zero to a vector has no effect: x+0 = 0+x = x.
 - Subtracting a vector from itself as long as the size and shape are the same yields a zero vector(x-x = 0)

```python
# Adding the vectors
sum = x1 + y1
print("Sum:",sum)

# OR

x_add = np.add(x1,y1)
print("Added vectors:",x_add)


# Subtracting the vector
sub = x1 - y1
print("Difference:",sub)

# OR

x_sub = np.subtract(x1,y1)
print("Subtracted vectors:",x_sub)
```

**Advanced Vector Operations: Dot Product (Scalar Product)**

Why This Matters: **Text Classification**

- Dot products are fundamental for text classification in **Natural Language Processing (NLP)**. By calculating dot products between word vectors and predefined sentiment vectors, you can determine the sentiment of a text. This enables automated sentiment analysis for large volumes of text data.

Given two vectors x and y in R^n(n- dimensional real number space), the dot product is defined as:

\begin{split} \mathbf{x} \cdot \mathbf{y} = \sum_{i=1}^{n} x_i y_i
  \end{split}



**Geometric Interpretation**

The dot product can also be expressed as:
\begin{split} \mathbf{x} \cdot \mathbf{y} = |\mathbf{x}| |\mathbf{y}| \cos(\theta)\end{split}
where:
-  (|x|) is the magnitude of vector(x)
-  (|y|) is the magnitude of vector(y)
-  (θ) is the angle between (x) and (y)

Properties:
1.  Commutative:(x.y=y.x)
2.  Distributes over vector addition:(x.(y+a)=x.y+x.a)
3.  Bilinear:(a(x.y)=(ax).y=x.(ay)),where (a) is a scaler

Applications:
-  Determining Orthogonality: Two vectors are orthogonal (perpendicular) if their dot product is zero.

-  Projection: Projecting one vector onto another.

- Finding Angles: The angle between two vectors can be determined using the dot product formula.



```python
# Let’s calculate the dot product of two vectors
x1=[1,3,-5]
y1=[4,-2,-1]
dot_product = np.dot(x1,y1)
print("Dot Products: ", dot_product)
```

**Advanced Vector Operations: Cross Product (Vector Product)**

Objective: Learn the dot and cross product of vectors.

- Why This Matters: 3D Graphics and Physics

    Cross products are crucial in 3D graphics and physics, particularly for calculating normals to surfaces, which are essential for lighting and rendering. They are also used in physics for finding torques and rotational vectors, helping to understand the dynamics of rotating bodies.

The cross product (also known as vector product) is an operation on two vectors in three-dimensional space, resulting in a vector that is perpendicular to both of the original vectors.

The cross product is defined as:

 \begin{split}
\mathbf{x} ×  \mathbf{y} = \begin{vmatrix}
\mathbf{i} & \mathbf{j} & \mathbf{k} \\
x_1 & x_2 & x_3 \\
y_1 & y_2 & y_3
\end{vmatrix}
\end{split}

 where
(i,j,k) are the unit vectors along the x, y, and z axes, and
(a1,a2.a3) and
(b1,b2,b3) are the components of vectors
(x) and
(y), respectively.


**Geometric Interpretation**

The magnitude of the cross product is given by:

\begin{split}|x|×|y|=|\mathbf{x}| |\mathbf{y}| \sin(\theta)\end{split}

where:
-  (|a|) an (|b|) are the magnitudes of vectors (x) and (y)
- (θ) is the angle between (x) and (y)

The direction of (x.y) is perpendicular to the plane formed by (x) and (y), determined by the right hand rule.

Properties:
1.  Non-Commutative:(x × y=-(y × x))
2. Distributed over vector addition:(x×(y+c)=x × y+x × c)
3. Not associative:(x × (y × c)!=(x × y) × c)

Applications
-  Normal Vector Calculation: Computing a vector perpendicular to a plane in 3D space.

- Torque: In physics, the cross product is used to determine the torque exerted by a force.

- Rotational Vectors: Understanding the rotation of objects in three-dimensional space.

```python
# Let’s calculate the cross product of two vectors
x1 = [1,3,-5]
y1 = [4,-2,-1]

cross_product = np.cross(x1,y1)
print("Cross Product: ",cross_product)
```

Activity: Vector Field Visualization for Beginners

Objective: Create a simple 2D vector field visualization using Python. Duration: Approximately 20 minutes Tools Required: Python, basic Python syntax knowledge.

```python
# Step 1: Import necessary libraries
import numpy as np
import matplotlib.pyplot as plt

# Step 2: Create Grid
x = np.linspace(-5,5,10)
y = np.linspace(-5,5,10)
# plt.xticks(np.arange(-5, 10.1, 1.1))
# plt.yticks(np.arange(-5, 10.1, 1.1))



X,Y = np.meshgrid(x,y)

# print(X)

# Step 3: Define the vector field
# U and V are the component of the vector
U = -Y  #Negaitve Y component for circular pattere
V = X # X componet for circular pattern

# Step 4: Plot the vector field
# plt.quiver(X,Y,U,V)
plt.quiver(X,Y,U,V)


plt.xlim(-5,5)
plt.ylim(-5,5)
plt.title("Circular Vector Field")
plt.xlabel("X axis")
plt.ylabel("Y axis")
plt.grid()

# Step 5 : Display the plot
plt.show()
```

\begin{split}
  X = \begin{bmatrix}
  x_{11} & x_{12} & \cdots & x_{1n} \\
  x_{21} & x_{22} & \cdots & x_{2n} \\
  \vdots & \vdots & \ddots & \vdots \\
  x_{m1} & x_{m2} & \cdots & x_{mn}
  \end{bmatrix}
  \end{split}
  where
 i is the number of rows, and j

**Matrix Multiplication: The Core of Complex Calculations**
-    A matrix of any shape can be multiplied by a scalar (i.e. a number describing magnitude).

Definition: Matrix multiplication, denoted as (Z = X.Y
), is an operation between two matrices(X
) ((i x j
)) and (Y
) ((k x l
)), where the number of columns in(X
) is equal to the number of rows in(Y
). The resulting matrix (C
) has dimensions ((m x n
)).

Mathematical Formula: The element (Z[i][j]
) of the resulting matrix (Z
) is obtained by taking the dot product of the(i
)th row of matrix(X
) and the (l
) column of matrix
(Y):

\begin{split}
Z[i][j] = \sum_{k=1}^{n} (X[i][k] \cdot Y[k][j])\end{split}

Key Rules:

1. Compatibility: For matrix multiplication to be defined, the number of columns in (X
) must be equal to the number of rows in
(Y) (i.e., the inner dimensions must match).

2. Associativity: Matrix multiplication is associative, meaning that
((X.Y).Z=X.(Y.Z)) if the dimensions allow.

3. Distributivity: Matrix multiplication distributes over matrix addition, i.e.,
(X.(Y+Z)=(X.Y)+(X.Z)).


Matrix Dot Product (Element-wise Product):
Definition: The matrix dot product, denoted as (Z = X.Y)
 is an element-wise operation between two matrices (A
) and (B
) of the same dimensions. The resulting matrix (Z
) has the same dimensions as (X) and (Y).

Mathematical Formula: Each element (Z[i][j]
) of the resulting matrix (Z
) is obtained by multiplying the corresponding elements of matrices
(X) and
(Y):

Z[i][j]=X[i][j]⋅Y[i][j]

Key Rule: Matrix dot product is applied element-wise, meaning that each element of the resulting matrix is calculated independently based on the corresponding elements of the input matrices.

Use Cases: Matrix dot product is often used in element-wise operations, such as calculating element-wise differences or similarities between matrices. It is commonly used in various mathematical and computational tasks, including image processing and certain types of data transformations.
